```python
# Task 01- :Data Preparation
import pandas as pd
import numpy as np
```

```python
# load the dataset
df = pd.read_csv('Telco_Customer_Churn_Dataset.csv')
```

```python
null = df.isnull().sum()
print(null)
```

```python
# now lets clean column names
df.columns = df.columns.str.strip()
```

```python
# Handle missing values
# Converting 'TotalCharges' to numeric, replacing empty strings with NaN
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
row_6754 = df.iloc[6754]
print(row_6754)
```

```python
# Filling missing 'TotalCharges' with median value
df['TotalCharges'].fillna(df['TotalCharges'].median(), inplace=True)
```

# Encoding Categorical variables




```python
# Identifying categorical columns
categorical_columns = df.select_dtypes(include=['object']).columns
print(categorical_columns)
```

```python
# drop customerID column
df = df.drop('customerID', axis=1)
```

```python
categorical_columns = df.select_dtypes(include=['object']).columns
print(categorical_columns)
```

```python
df.head()
```

```python
# applying labelEncoder to binary categorical variables
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
binary_columns = ['gender', 'Partner', 'Dependents', 'PhoneService', 'PaperlessBilling', 'Churn']
for col in binary_columns:
    df[col] = le.fit_transform(df[col])
```

```python
df.head()
```

```python
print(df.info())
```

```python
# one hot encoding to multi-category variables
multi_category_columns = [col for col in categorical_columns if col not in binary_columns]
df = pd.get_dummies(df, columns=multi_category_columns)
```

```python
print(df.info())
```

```python
# save the preprocessed dataset to a new CSV File
df.to_csv('Processed_Telco_Customer_Churn.csv', index=False)
print("\nProcessed dataset saves as 'Processed_Telco_Customer_Churn.csv'")
```
