```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
```

```python
# Defining candidate models with initial parameters
models = {
    'Logistic Regression': LogisticRegression(
        max_iter=1000,
        random_state=42,
        class_weight='balanced'  # To handle class imbalance
    ),
    'Random Forest': RandomForestClassifier(
        n_estimators=100,
        random_state=42,
        class_weight='balanced'  # To handle class imbalance
    ),
    'XGBoost': XGBClassifier(
        use_label_encoder=False,
        eval_metric='logloss',
        random_state=42,
        scale_pos_weight=len(y_train[y_train == 0]) / len(y_train[y_train == 1])  # To handle class imbalance
    )
}
```

```python
# Printing model selection details
print("Selected Models for Churn Prediction:")
for model_name, model in models.items():
    print(f"\nModel: {model_name}")
    print(f"Parameters: {model.get_params()}")
    print(f"Justification:")
    if model_name == 'Logistic Regression':
        print(" - Simple and interpretable, suitable for binary classification.")
        print(" - Provides feature importance via coefficients.")
        print(" - Class weight set to 'balanced' to address churn dataset imbalance.")
    elif model_name == 'Random Forest':
        print(" - Robust ensemble model that captures non-linear relationships.")
        print(" - Less sensitive to feature scaling and handles imbalanced data well.")
        print(" - Provides feature importance for interpretability.")
    elif model_name == 'XGBoost':
        print(" - Powerful gradient boosting model, often performs well on imbalanced datasets.")
        print(" - Handles class imbalance via scale_pos_weight parameter.")
        print(" - Captures complex patterns and interactions in the data.")
```

# Model Evaluation

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, roc_curve
```

```python
# Initializing a dictionary to store evaluation metrics
evaluation_results = {
    'Model': [],
    'Accuracy': [],
    'Precision': [],
    'Recall': [],
    'F1-Score': [],
    'ROC-AUC': []
}
```

```python
# plot setup for ROC Curves
plt.figure(figsize=(10, 6))
```

```python
# Evaluating each model
for model_name, model in models.items():
    print(f"\nEvaluating {model_name}...")
    
    # Making predictions on the test set
    y_pred = model.predict(x_test_selected)
    y_pred_proba = model.predict_proba(x_test_selected)[:, 1]  # Probability for positive class (Churn=1)
    
    # Calculating metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_pred_proba)
    
    # Storing metrics
    evaluation_results['Model'].append(model_name)
    evaluation_results['Accuracy'].append(accuracy)
    evaluation_results['Precision'].append(precision)
    evaluation_results['Recall'].append(recall)
    evaluation_results['F1-Score'].append(f1)
    evaluation_results['ROC-AUC'].append(roc_auc)
    
    # Printing confusion matrix
    print(f"\nConfusion Matrix for {model_name}:")
    cm = confusion_matrix(y_test, y_pred)
    print(cm)
    
    # Visualizing confusion matrix
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.title(f'Confusion Matrix - {model_name}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()
    
    # Calculating and plotting ROC curve
    fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
    plt.plot(fpr, tpr, label=f'{model_name} (AUC = {roc_auc:.2f})')

# Finalizing ROC curve plot
plt.plot([0, 1], [0, 1], 'k--')  # Diagonal line for random guessing
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves for Churn Prediction Models')
plt.legend(loc='lower right')
plt.show()

# Creating a DataFrame for evaluation results
results_df = pd.DataFrame(evaluation_results)

# Displaying the evaluation results
print("\nModel Evaluation Results:")
print(results_df)

# Analysis of best-performing model
best_model = results_df.loc[results_df['F1-Score'].idxmax()]
print(f"\nBest Model Based on F1-Score: {best_model['Model']}")
print(f"Metrics for {best_model['Model']}:")
print(f"Accuracy: {best_model['Accuracy']:.4f}")
print(f"Precision: {best_model['Precision']:.4f}")
print(f"Recall: {best_model['Recall']:.4f}")
print(f"F1-Score: {best_model['F1-Score']:.4f}")
print(f"ROC-AUC: {best_model['ROC-AUC']:.4f}")
print("\nAnalysis:")
print(f"The {best_model['Model']} model performs best based on F1-Score, which balances precision and recall, making it suitable for imbalanced datasets like churn prediction.")
print("Considerations for improvement: Tune hyperparameters, address class imbalance further, or try additional models if performance is suboptimal.")
```

# Model Deployment Preparation

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
import joblib

# Loading the original dataset
try:
    df = pd.read_csv('../Datasets/Telco_Customer_Churn_Dataset.csv')
except FileNotFoundError:
    print("Error: 'Telco_Customer_Churn_Dataset.csv' not found. Please upload the file to Colab.")
    raise
```

```python
# Debugging: Print column names to verify
print("\nOriginal DataFrame columns:", df.columns.tolist())
```

```python
# Preprocess steps
# Dropping 'customerID' if it exists
if 'customerID' in df.columns:
    df = df.drop('customerID', axis=1)
else:
    print("Warning: 'customerID' not found in DataFrame.")
```

```python
# Handling missing values in 'TotalCharges'
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df['TotalCharges'].fillna(df['TotalCharges'].median(), inplace=True)
```

```python
# Debugging: Print column names after preprocessing
print("\nDataFrame columns after preprocessing:", df.columns.tolist())
```

```python
# Identifying categorical and numerical columns
categorical_columns = df.select_dtypes(include=['object']).columns
# Expected binary columns (adjust if your dataset uses different names)
expected_binary_columns = ['gender', 'Partner', 'Dependents', 'PhoneService', 'PaperlessBilling']
binary_columns = [col for col in expected_binary_columns if col in df.columns]
if len(binary_columns) < len(expected_binary_columns):
    print("Warning: Some expected binary columns not found:", set(expected_binary_columns) - set(binary_columns))

multi_category_columns = [col for col in categorical_columns if col not in binary_columns and col != 'Churn']
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns

# Debugging: Print identified columns
print("\nBinary columns:", binary_columns)
print("Multi-category columns:", multi_category_columns)
print("Numerical columns:", numerical_columns)
```

```python
# Defining the preprocessing pipeline
preprocessor = ColumnTransformer(
    transformers=[
        ('binary', OrdinalEncoder(), binary_columns),  # Use OrdinalEncoder for binary columns
        ('multi_cat', OneHotEncoder(drop='first', sparse_output=False, handle_unknown='ignore'), multi_category_columns),
        ('num', 'passthrough', numerical_columns)  # Numerical columns pass through unchanged
    ])
```

```python
# Creating the full pipeline
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression(max_iter=1000, random_state=42, class_weight='balanced'))
])
```

```python
# Fitting the pipeline
try:
    X_train_full = df.drop('Churn', axis=1)
    y_train_full = df['Churn'].apply(lambda x: 1 if x == 'Yes' else 0)  # Converting 'Churn' to binary
    pipeline.fit(X_train_full, y_train_full)
    print("\nPipeline fitted successfully.")
except KeyError as e:
    print(f"Error: {e}. Please check if 'Churn' is in the DataFrame columns:", df.columns.tolist())
    raise
```

```python
# Step 4: Saving the pipeline
joblib.dump(pipeline, '../Trained_models/Logistic_regression_pipeline_fixed.pkl')
print("\nLogistic Regression pipeline saved as 'Logistic_regression_pipeline_fixed.pkl'")
```

```python
# Function to predict churn
def predict_churn(new_data, pipeline_path='../Trained_models/Logistic_regression_pipeline_fixed.pkl'):
    
    # Loading the pipeline
    pipeline = joblib.load(pipeline_path)
    
    # Ensuring new_data is a DataFrame
    if isinstance(new_data, dict):
        new_data = pd.DataFrame([new_data])
    
    # Storing customerID if present
    customer_ids = new_data['customerID'] if 'customerID' in new_data.columns else None
    
    # Dropping 'customerID' if present
    if 'customerID' in new_data.columns:
        new_data = new_data.drop('customerID', axis=1)
    
    # Ensuring new_data has the same columns as the original dataset (excluding 'Churn')
    expected_columns = df.drop('Churn', axis=1).columns
    new_data = new_data.reindex(columns=expected_columns, fill_value=0)
    
    # Making predictions
    predictions = pipeline.predict(new_data)
    probabilities = pipeline.predict_proba(new_data)[:, 1]
    
    # Formatting results
    results = {
        'customerID': customer_ids.tolist() if customer_ids is not None else [f'Customer_{i+1}' for i in range(len(predictions))],
        'Predicted_Churn': ['Yes' if pred == 1 else 'No' for pred in predictions],
        'Churn_Probability': probabilities
    }
    
    return pd.DataFrame(results)
```

```python
# Testing predictions on sample data
# Example 1: Single new data point
single_data = {
    'customerID': '1234-XYZ',
    'gender': 'Male',
    'SeniorCitizen': 0,
    'Partner': 'No',
    'Dependents': 'No',
    'tenure': 12,
    'PhoneService': 'Yes',
    'MultipleLines': 'No',
    'InternetService': 'DSL',
    'OnlineSecurity': 'No',
    'OnlineBackup': 'No',
    'DeviceProtection': 'No',
    'TechSupport': 'No',
    'StreamingTV': 'No',
    'StreamingMovies': 'No',
    'Contract': 'Month-to-month',
    'PaperlessBilling': 'Yes',
    'PaymentMethod': 'Electronic check',
    'MonthlyCharges': 50.0,
    'TotalCharges': 600.0
}

print("\nPrediction for Single New Data Point:")
try:
    single_result = predict_churn(single_data)
    print(single_result)
except Exception as e:
    print(f"Error predicting single data: {e}")
```
